```python
def factorial(number: int) -> int:
    result = number
    for i in range(2,number):
        result = result * i % 100
    return result
```

```python
def dec_and_digit(number: int) -> tuple[int,int]:
    number %= 100
    digit = number // 10
    dec = number % 10
    return (digit,dec)
```

```python
def dec_and_digit2(number: int) -> tuple[int,int]:
    digit = number % 10
    dec = number % 100 // 10
    return (dec,digit)
```

```python
for x in range(1,100):
    print(dec_and_digit(x), "-", dec_and_digit2(x))
```
